```python

import requests
from bs4 import BeautifulSoup
#解析网页，并爬取城市天气数据，生成列表
def parse_page(url):
    #头文件
    headers = {
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                     'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36'
    }
    response = requests.get(url,headers=headers)
    #解决编码问题
    text = response.content.decode('utf-8')
    soup = BeautifulSoup(text,'lxml')
    #解析网页：定位 conMidtab 标签
    conMidtab = soup.find('div',class_ = 'conMidtab')
    #print(conMidtab)#打印输出看输出是否符合条件
    tables = conMidtab.find_all('table')
    #查看是否拿到了每个城市的天气
     #info存放城市天气信息
    info = []
    for table in tables:
        trs = table.find_all('tr')[2:]
        provence = []
        for tr in trs:
            cityinfo = []
            tdc = tr.find_all('td')
            for i in tdc:
                #将标签中的信息添加到城市信息列表中
                cityinfo.append(list(i.stripped_strings))
            #根据省添加城市
            provence.append(cityinfo)
        #将省信息添加在总列表中
        info.append(provence)
    #整理list，将每个城市从省列表中提到城市列表中
    cityList = []
    for first in info:
        for second in first:
            #print(len(second))
            if len(second) == 9:
                cityList.append(second[1:])
            elif len(second) == 8:
                cityList.append(second)
    return cityList
#-------------------------------------------------------------------------------------------------------
#主要功能：在城市列表中获取目标城市
def foundCity(cityName,cityList):
    #目标城市元素字典
    displayInfo = {}
    for i in cityList:
#         print(i[0])
        if i[0][0] == cityName:
            displayInfo = {'城市':i[0][0],'最高气温':i[3][0],'最低气温':i[6][0],'日间天气':i[1][0],'日间风向风力':i[2][0],'夜间天气':i[4][0],'夜间风向风力':i[5][0],'最低气温':i[6][0]}
#     print(displayInfo)
    return displayInfo

#-------------------------------------------------------------------------------------------------------
#输出控制
def printInfo(displayInfo,date):
    print('===============================================')
    print(date,str(list(displayInfo.values())[0])+'天气情况')
    print('-------------------------------------------')
    for item in range(1,len(displayInfo)):

        print(('-'*2+list(displayInfo.keys())[item]+':'+list(displayInfo.values())[item]))
#         print('%s : %s'%(str(displayInfo.keys()[item]),str(displayInfo.values()[item])))
#         print(list(displayInfo.keys))
#         print(list(displayInfo.keys()))
#         print(displayInfo.keys())
#         print(list(displayInfo.values))
#     return displayInfo
    print('===============================================')

#-------------------------------------------------------------------------------------------------------
#主程序
def main():
    urls = [
        'http://www.weather.com.cn/textFC/hb.shtml',
        'http://www.weather.com.cn/textFC/db.shtml',
        'http://www.weather.com.cn/textFC/hz.shtml',
        'http://www.weather.com.cn/textFC/hn.shtml',
        'http://www.weather.com.cn/textFC/hd.shtml',
        'http://www.weather.com.cn/textFC/xb.shtml',
        'http://www.weather.com.cn/textFC/xn.shtml',
        'http://www.weather.com.cn/textFC/gat.shtml'
    ]
    cityName = input('Please input the city name what you want to search (例：海南):')
    date = input('Please input the date what you want to search (例：2019-05-06):')
    cityAll = []
    for url in urls:
        #print(cityList)
        for i in parse_page(url):
            cityAll.append(i[:])
    #print(cityAll)
    displayInfo = foundCity(cityName,cityAll)
    printInfo(displayInfo,date)


#----------------------------------------
```
